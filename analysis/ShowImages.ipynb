```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

from pathlib import Path
from astropy.nddata import CCDData
from astropy.io import fits
from ccdproc import ImageFileCollection
import ccdproc as ccdp

from astropy.nddata import CCDData
from astropy.stats import mad_std
import astropy.units as u
from convenience_functions import show_image

# suppress warnings is not recommended :I
import warnings
warnings.filterwarnings("ignore")

mpl.rcParams['figure.dpi'] = 200
```

```python
cc_com_reduced = './data/cc_com/reduced/'
tz_boo_reduced = './data/tz_boo/reduced/'

def display_processed(images_dir, figsize=(20, 12), cmap='gray'):
    images = ImageFileCollection(images_dir)
    images = images.files_filtered(include_path=True)
    fig, axs = plt.subplots(1, len(images), figsize=figsize);
    
    for i, path in np.ndenumerate(images):
        image_data = CCDData.read(path, unit='adu').data
        print(np.median(image_data))
        show_image(image_data, fig=fig, ax=axs[i], cmap=cmap);
        
    fig.tight_layout();
        
    return fig, axs
```

```python
display_processed(cc_com_reduced)
```
